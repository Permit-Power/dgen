```python
import pandas as pd

# --- Configuration ---
RESULTS_DIR     = "/Users/wael/Documents/dgen_runs"   # directory containing the xlsx file
FILE_NAME       = "original_md_policy_run.xlsx"           # filename written by analysis_of_model_results
STATE           = "MD"                 # state abbreviation
SYSTEM_LIFETIME = 25                   # years

# --- Load all sheets ---
xlsx_path = f"{RESULTS_DIR}/{FILE_NAME}"
sheets = pd.read_excel(xlsx_path, sheet_name=None)  # dict of {sheet_name: DataFrame}

print("Sheets loaded:", list(sheets.keys()))
```

```python
df_eabs = sheets["eabs_states"]
df_pas = sheets["portfolio_annual_savings"]


lower_net_bill_savings = (
    df_eabs.loc[
        (df_eabs["scenario"] == "policy") & (df_eabs["year"] == 2040),
        "eabs_net"
    ].iloc[0]
)


lifetime_net_bill_savings = (
    df_pas.loc[
        (df_pas["scenario"] == "policy") & (df_pas["year"] == 2040),
        "median_lifetime_net_savings_per_adopter_to_date"
    ].iloc[0]
)

current_net_bill_savings_fmt = f"${lower_net_bill_savings / 1000:.2f}k"
lifetime_net_bill_savings_fmt       = f"${lifetime_net_bill_savings / 1000:.0f}k"

print(f"current_net_bill_savings: {current_net_bill_savings_fmt}/yr")
print(f"lifetime_net_bill_savings:       {lifetime_net_bill_savings_fmt}")
```

```python
df_totals = sheets["totals"]

policy_installs_2040 = (
    df_totals.loc[
        (df_totals["scenario"] == "policy") & (df_totals["year"] == 2040),
        "number_of_adopters"
    ].iloc[0]
)


baseline_installs_2040 = (
    df_totals.loc[
        (df_totals["scenario"] == "baseline") & (df_totals["year"] == 2040),
        "number_of_adopters"
    ].iloc[0]
)


percent_increase_installs = f"{(policy_installs_2040 / baseline_installs_2040 - 1)*100:.0f}%"
number_increase_installs = f"{(policy_installs_2040 - baseline_installs_2040):.0f}"

print(f"Percent increase in installs by 2040: {percent_increase_installs}")
print(f"Number increase in installs by 2040: {number_increase_installs}")
```
